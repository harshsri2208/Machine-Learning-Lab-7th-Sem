```python
# getting pixel values for each point in river_loc

river_val = []

for i in range(len(river_loc)) :
    
    river_val_row = []
    x = river_loc[i][0]
    y = river_loc[i][1]
    
    river_val_row.append(pix1[x, y][0])
    river_val_row.append(pix2[x, y][0])
    river_val_row.append(pix3[x, y][0])
    river_val_row.append(pix4[x, y][0])
    
    river_val.append(river_val_row)
    
print(river_val)
```

```python
# getting pixel values for each point in non_river_loc

non_river_val = []

for i in range(len(non_river_loc)) :
    
    non_river_val_row = []
    x = non_river_loc[i][0]
    y = non_river_loc[i][1]
    
    non_river_val_row.append(pix1[x, y][0])
    non_river_val_row.append(pix2[x, y][0])
    non_river_val_row.append(pix3[x, y][0])
    non_river_val_row.append(pix4[x, y][0])
    
    non_river_val.append(non_river_val_row)
    
print(non_river_val)
```

```python
# coverting to np arrays

river_val = np.array(river_val)
non_river_val = np.array(non_river_val)

print(river_val.shape)
print(non_river_val.shape)
```

```python
# mean of river class

T1 = np.mean(river_val, axis = 0)
T1
```

```python
# mean of non river class

T2 = np.mean(non_river_val, axis = 0)
T2
```

```python
# calculating deviation from mean for river and non river classes

deviation_river = river_val - T1
deviation_non_river = non_river_val - T2

# calculating the transpose for the matrices obtaained above

deviation_river_T = np.transpose(deviation_river)
deviation_non_river_T = np.transpose(deviation_non_river)

print(deviation_river)
print(deviation_non_river.shape)
print(deviation_river_T.shape)
print(deviation_non_river_T.shape)
```

```python
# covariance matrix for river class

cov_river = np.dot(np.transpose(deviation_river), deviation_river) / 50.0

print(cov_river)
```

```python
# covariance matrix for non river class

cov_non_river = np.dot(np.transpose(deviation_non_river), deviation_non_river) / 100.0

print(cov_non_river)
```

```python
# inverse of covariance matrices
inv_river_cov = np.linalg.inv(cov_river)
inv_non_river_cov = np.linalg.inv(cov_non_river)

# determinant of covariance matrices
det_river_cov = np.linalg.det(cov_river)
det_non_river_cov = np.linalg.det(cov_non_river)
```

```python
# count variable to provide diffrent names to output files for 3 cases of Prior Probabilities

count = 1
```

```python
# function implementation of Bayes Decision Rule
# takes Prior probablities of two classes as input
# formaula for multivariate gaussian distribution is used
# probablity values for both classes are calculated
# now if P1*p1 >= P2*p2 then the point is classified as river and RGB values is set to (255, 255, 255)
# else RGB value is set to (0, 0, 0)


def final_output(P1, P2) : 
    
    global count
    
    output_img = Image.new('RGB', (676, 598))
    output_img.save('blank.jpg')
    pix_output = output_img.load()
    
    for i in range(676) :
        for j in range(598) :

            test_data = []

            test_data.append(pix1[i, j][0])
            test_data.append(pix2[i, j][0])
            test_data.append(pix3[i, j][0])
            test_data.append(pix4[i, j][0])

            test_data = np.array(test_data)

            test_dev_river = test_data - T1
            river_class = np.transpose(test_dev_river).dot(inv_river_cov.dot(test_dev_river))

            test_dev_non_river = test_data - T2
            non_river_class = np.transpose(test_dev_non_river).dot(inv_non_river_cov.dot(test_dev_non_river))

            p1 = np.exp((-0.5) * river_class) / np.sqrt(det_river_cov)
            p2 = np.exp((-0.5) * non_river_class) / np.sqrt(det_non_river_cov)

            if (P1 * p1) >= (P1 * p2) :
                pix_output[i, j] = (255, 255, 255)
            else:
                pix_output[i, j] = (0, 0, 0)

    final_file = 'output' + (str(count)) + '.jpg'
    output_img.save(final_file)

    img=mpimg.imread(final_file) # similarly we can see other images
    imgplot = plt.imshow(img)
    plt.show()
    
    count = count + 1 # creating different images for cases
```

```python
# for P1 = 0.3 AND P2 = 0.7
final_output(0.3, 0.7)
```
